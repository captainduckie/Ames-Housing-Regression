```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Function for comparing different approaches
def score_dataset(X, y, n_estimators):
  
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

    model = RandomForestRegressor(n_estimators=n_estimators, random_state=0)
    model.fit(train_X, train_y)
    preds = model.predict(val_X)
    rmse = mean_squared_error(val_y, preds, squared=False)
    return rmse
```

```python
from sklearn.impute import SimpleImputer

# Get names of columns with missing values
cols_with_missing = [col for col in OH_X.columns
                     if OH_X[col].isnull().any()]

# Make copy to avoid changing original data (when imputing)
OH_X_plus = OH_X.copy()

# Make new columns indicating what will be imputed
for col in cols_with_missing:
    OH_X_plus[col + '_was_missing'] = OH_X[col].isnull()

# Imputation
my_imputer = SimpleImputer()
imputed_OHX_plus = pd.DataFrame(my_imputer.fit_transform(OH_X_plus))

# Imputation removed column names; put them back
imputed_OHX_plus.columns = OH_X_plus.columns
```

# Tuning Random forest hyperparameters

```python
def get_mse(n_estimators, max_leaf_nodes, X, y):

    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

    model = RandomForestRegressor(n_estimators=n_estimators, max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)

    mse = mean_squared_error(val_y, preds_val, squared=False)

    return mse
```

```python
# define function to compare and return optimal n_estimators and max_leaf_nodes using MAE
def optimize_parameters_rf(n_estimators_ls, max_leaf_nodes_ls, X, y):

  # initialize variables
  best_n_estimators = 0
  best_max_leaf_nodes = 0
  lowest_rmse = 1000000

  for n_estimators in n_estimators_ls:
      for max_leaf_nodes in max_leaf_nodes_ls:
        my_rmse= get_mse(n_estimators, max_leaf_nodes, X, y)

        if my_rmse < lowest_rmse:
          lowest_rmse = my_rmse
          best_n_estimators = n_estimators
          best_max_leaf_nodes = max_leaf_nodes
  
  print("OPTIMAL PARAMETERS | n estimators: %d  \t\t max_leaf_nodes: %d  \t\t RMSE:  %d" %(best_n_estimators, best_max_leaf_nodes, lowest_rmse))
```

```python
# compare MAE with differing values of n_estimators and max_leaf_nodes
n_estimators_ls = [1,10,100,500,1000]
max_leaf_nodes_ls = [5,50,500,1000]

optimize_parameters_rf(n_estimators_ls, max_leaf_nodes_ls, imputed_OHX_plus, y)
```
